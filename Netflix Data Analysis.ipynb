```python
#3) Which countries contribute the most to the content library?
# Check for missing values in the 'country' column
print(data['country'].isnull().sum())

# Fill missing values with 'Unknown' (or another appropriate value)
data['country'] = data['country'].fillna('Unknown')

# Count the occurrences of each country
country_counts = data['country'].value_counts()

# Print the top N countries (e.g., top 10)
top_n = 10
print(country_counts.head(top_n))
```

```python
#1) Find out which types of genre has most movies and TV Shows
movgen=data['release_year'].value_counts()
print(movgen)
mostmov=movgen.index[0]
print(mostmov)
```

```python
#5) What are the most common ratings, and how are they distributed across different types of content?
comrat=data['rating'].value_counts().index[0]
print(comrat)
plt.figure(figsize=(12,6))
sns.countplot(data=data,x='rating',hue='type')
```

```python
#2) What is the distribution of content types (e.g., Movies vs. TV Shows) in the dataset?
data['type'].value_counts()
sns.countplot(x='type',data=data)
```

```python
#4) What is the average duration of movies, and how does it vary across different ratings

# Calculate the average duration of movies
average_duration = data['duration'].mean()
print(f"The average duration of movies is: {average_duration}")

# Analyze duration variation across different ratings
rating_duration = data.groupby('rating')['duration'].mean()
print("\nAverage duration by rating:\n", rating_duration)

# Visualization (optional)
plt.figure(figsize=(10, 6))
sns.barplot(x=rating_duration.index, y=rating_duration.values)
plt.xlabel("Rating")
plt.ylabel("Average Duration")
plt.title("Average Movie Duration by Rating")
plt.show()
```
