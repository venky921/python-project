```python
#1) What is the overall distribution of sentiments (positive, negative, neutral) across the posts?
#2) Which platform generates the most engagement in terms of likes and retweets?
#3) Which users have the highest average engagement (likes and retweets per post)?
dist=data['Sentiment'].value_counts()
print(dist)
plt.figure(figsize=(192,96))
data=data
sns.countplot(data,x='Sentiment',)
```

```python
#2) Which platform generates the most engagement in terms of likes and retweets?

# Calculate total engagement for each platform (sum of likes and retweets)
data['Total_Engagement'] = data['Likes'] + data['Retweets']

# Group by platform and calculate the sum of total engagement
platform_engagement = data.groupby('Platform')['Total_Engagement'].sum()

# Find the platform with the highest engagement
most_engaged_platform = platform_engagement.idxmax()

print(f"The platform with the most engagement (likes + retweets) is: {most_engaged_platform}")
```

```python
#3) Which users have the highest average engagement (likes and retweets per post)?
# Calculate the total engagement (likes + retweets) for each user
data['TotalEngagement'] = data['Likes'] + data['Retweets']

# Group by user and calculate the average engagement
average_engagement = data.groupby('User')['TotalEngagement'].mean()

# Find the users with the highest average engagement
highest_engagement_users = average_engagement.nlargest(5) # Get the top 5 users

highest_engagement_users
```
